# Análise exploratória das principais variáveis


Durante a etapa inicial de preparação dos dados, optou-se por realizar um conjunto de transformações com o objetivo de simplificar a base e eliminar variáveis redundantes ou pouco informativas para o modelo de regressão linear múltipla. Essa redução buscou evitar problemas de multicolinearidade e manter o foco em atributos com potencial relação direta com o desempenho acadêmico.

Primeiramente, foi removida a variável **school**, uma vez que a base contém apenas duas escolas e não há interesse analítico em comparar instituições — o objetivo central é compreender o desempenho geral dos estudantes. Também foram excluídas as variáveis **Mjob** e **Fjob**, referentes às ocupações dos pais. Embora possam refletir o status socioeconômico, tais variáveis são altamente categóricas e pouco homogêneas entre os alunos, o que dificultaria sua utilização direta na regressão linear sem codificações complexas.

As variáveis **reason** (motivo da escolha da escola) e **guardian** (responsável legal) foram suprimidas por representarem aspectos contextuais ou institucionais menos diretamente associados ao desempenho escolar. Da mesma forma, **romantic** (relação amorosa) e **famrel** (qualidade do relacionamento familiar) foram removidas, pois, apesar de potencialmente relevantes sob uma ótica sociológica, sua subjetividade e dificuldade de mensuração consistente poderiam introduzir ruído ao modelo.

Além das exclusões, realizou-se a **aglutinação das variáveis de educação parental** — **Medu** (educação materna) e **Fedu** (educação paterna) — em uma nova variável chamada **ParEdu**, calculada como a média aritmética das duas. Essa transformação visa capturar de maneira mais estável o nível médio de escolaridade dos responsáveis, um indicador robusto do capital educacional familiar. Embora a base original não contenha uma variável explícita de renda, a variável ParEdu também pode atuar como um proxy aproximado para esse aspecto socioeconômico.

Essas modificações resultaram em uma tabela mais enxuta, mantendo apenas as variáveis com maior potencial explicativo e interpretativo para a previsão do desempenho final dos estudantes (G3).

# Algoritmo

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import statsmodels.stats.api as sms
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import RFE
from sklearn.metrics import r2_score, mean_squared_error
import seaborn as sns
import statsmodels.formula.api as smf
import os
from google.colab import drive


# ===============================================================
# ====================== 1. EXPLORAÇÂO DE DADOS =================
# ===============================================================

# ---------- 1.1 Carregar e pré-processar em ambiente collab ----------

drive.mount('/content/drive')
path = '/content/drive/MyDrive/MQAM/student-mat.csv'  #É necessário adaptar o path para o endereço equivalente ao endereço de diretório que a base de dados foi armazenada
df = pd.read_csv(path, sep=';', encoding= 'utf-8')

# Remover colunas irrelevantes e criar ParEdu
cols_to_drop = ["school", "Mjob", "Fjob", "reason", "guardian", "romantic", "famrel"]
df = df.drop(columns=cols_to_drop, errors="ignore")

if ("Medu" in df.columns) and ("Fedu" in df.columns):
    df["ParEdu"] = df[["Medu", "Fedu"]].mean(axis=1)
    df = df.drop(columns=["Medu", "Fedu"], errors="ignore")

target = "G3"
if target not in df.columns:
    raise RuntimeError("Variável resposta G3 não encontrada.")

# ---------- 1.2 Separar variáveis numéricas e categóricas ----------
num_vars = df.select_dtypes(include=[np.number]).columns.tolist()
cat_vars = df.select_dtypes(include=["object", "category"]).columns.tolist()

# ---------- 1.3 Codificação das categóricas ----------
df_encoded = pd.get_dummies(df, columns=cat_vars, drop_first=True)

# ---------- 1.4 Tratamento de missing values ----------
df_encoded = df_encoded.fillna(df_encoded.mean())

# ---------- 1.5 Divisão treino/teste ----------
X = df_encoded.drop(columns=[target])
y = df_encoded[target]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# ---------- 1.6 Padronização ----------
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# ---------- 1.7 Seleção de variáveis com RFE ----------
base_model = LinearRegression()
rfe = RFE(base_model, n_features_to_select=10)
rfe.fit(X_train_scaled, y_train)

selected_features = X_train.columns[rfe.support_]
print("\nVariáveis selecionadas pelo RFE:\n", selected_features.tolist())

# ---------- 1.8 Reajuste do modelo ----------
X_train_sel = X_train_scaled[:, rfe.support_]
X_test_sel = X_test_scaled[:, rfe.support_]

model = LinearRegression()
model.fit(X_train_sel, y_train)

y_pred = model.predict(X_test_sel)

# ---------- 1.9 Avaliação ----------
r2 = r2_score(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
cv_r2 = cross_val_score(model, X_train_sel, y_train, cv=5, scoring='r2').mean()

print(f"\nR² de teste: {r2:.3f}")
print(f"R² (cross-val média): {cv_r2:.3f}")
print(f"RMSE: {rmse:.3f}")

# ---------- 1.10 Coeficientes e importância ----------
coef_df = pd.DataFrame({
    "Variável": selected_features,
    "Coeficiente": model.coef_
}).sort_values(by="Coeficiente", ascending=False)
print("\nCoeficientes do modelo:\n", coef_df)
```

```python
# ===============================================================
# ====================== 2. MODELAGEM E TESTES ==================
# ===============================================================

# ---------- 2.1 Modelo OLS (básico) ----------
# Modelo inicial de Regressão Linear usando apenas variáveis numéricas
X = df[num_vars].drop(columns=[target], errors="ignore").fillna(df[num_vars].mean())
X = sm.add_constant(X)  # Inclui termo constante (intercepto)
y = df[target].fillna(df[target].mean())

model = sm.OLS(y, X).fit()
print("\n--- Modelo OLS inicial (todas as numéricas) ---")
print(model.summary())  # Exibe coeficientes, p-valores e métricas R², AIC etc.

# ---------- 2.2 Normalidade e heterocedasticidade ----------
# Avalia pressupostos do modelo linear: resíduos devem ser ~N(0, σ²)
resid = model.resid
fitted = model.fittedvalues

# Teste de Breusch-Pagan: detecta heterocedasticidade
# H0 = homocedasticidade (variância constante)
bp_test = sms.het_breuschpagan(resid, model.model.exog)
bp_labels = ["LM stat", "LM p-value", "F-stat", "F p-value"]
print("\nBreusch-Pagan:", dict(zip(bp_labels, bp_test)))
# Interpretação: p-value < 0.05 → rejeita H0 → presença de heterocedasticidade

# ---------- 2.3 Outliers (IQR) ----------
# Contagem de valores extremos via regra do intervalo interquartílico (IQR)
outliers = {}
for col in num_vars:
    Q1, Q3 = df[col].quantile(0.25), df[col].quantile(0.75)
    IQR = Q3 - Q1
    mask = (df[col] < (Q1 - 1.5 * IQR)) | (df[col] > (Q3 + 1.5 * IQR))
    outliers[col] = int(mask.sum())

print("\nNúmero de outliers por variável:")
display(pd.Series(outliers).sort_values(ascending=False))

# ---------- 2.4 Melhor modelo (RFE + avaliação) ----------
# Codificação de variáveis categóricas (one-hot) e substituição de NA por médias
df_encoded = pd.get_dummies(df, columns=cat_vars, drop_first=True)
df_encoded = df_encoded.fillna(df_encoded.mean(numeric_only=True))

X = df_encoded.drop(columns=[target])
y = df_encoded[target]

# Divisão treino/teste (75% / 25%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Padronização das variáveis (média 0, desvio 1)
scaler = StandardScaler()
X_train_s = scaler.fit_transform(X_train)
X_test_s = scaler.transform(X_test)

# Seleção recursiva de atributos (RFE)
rfe = RFE(LinearRegression(), n_features_to_select=10)
rfe.fit(X_train_s, y_train)
selected = X_train.columns[rfe.support_]

print("\nVariáveis selecionadas (RFE):", selected.tolist())

# Ajuste do modelo final
model_refined = LinearRegression().fit(X_train_s[:, rfe.support_], y_train)
y_pred = model_refined.predict(X_test_s[:, rfe.support_])

# Métricas de desempenho
r2 = r2_score(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
cv_r2 = cross_val_score(model_refined, X_train_s[:, rfe.support_],
                        y_train, cv=5, scoring='r2').mean()

print(f"\nR² (teste): {r2:.3f}")
print(f"R² (cross-val): {cv_r2:.3f}")
print(f"RMSE: {rmse:.3f}")
```

```python
# ===============================================================
# ========== 3. SELEÇÃO E INTERPRETAÇÃO DAS VARIÁVEIS ===========
# ===============================================================

def stepwise_selection(data, target, initial_list=[], threshold_in=0.05, threshold_out=0.10, verbose=True):
    """
    Função de seleção Stepwise baseada em p-valores (AIC-like) para regressão linear (OLS).
    Inclui tratamento automático de tipos e valores ausentes.
    """

    data = data.copy()

    # --- 3.1 Garantir que as colunas são numéricas (inclusive y) ---
    for col in data.columns:
        data[col] = pd.to_numeric(data[col], errors="coerce")

    # --- 3.2 Substituir valores ausentes por médias ---
    data = data.fillna(data.mean())

    # --- 3.3 Separar y e X ---
    y = data[target].astype(float)
    X = data.drop(columns=[target]).astype(float)

    included = list(initial_list)

    while True:
        changed = False

        # --- Etapa Forward ---
        excluded = list(set(X.columns) - set(included))
        new_pval = pd.Series(index=excluded, dtype=float)

        for new_column in excluded:
            model = sm.OLS(y, sm.add_constant(pd.DataFrame(X[included + [new_column]]))).fit()
            new_pval[new_column] = model.pvalues[new_column]

        best_pval = new_pval.min() if not new_pval.empty else None

        if best_pval is not None and best_pval < threshold_in:
            best_feature = new_pval.idxmin()
            included.append(best_feature)
            changed = True
            if verbose:
                print(f"✅ Adicionada variável {best_feature} com p-valor {best_pval:.3f}")

        # --- Etapa Backward ---
        if included:
            model = sm.OLS(y, sm.add_constant(pd.DataFrame(X[included]))).fit()
            pvalues = model.pvalues.iloc[1:]
            worst_pval = pvalues.max()

            if worst_pval > threshold_out:
                worst_feature = pvalues.idxmax()
                included.remove(worst_feature)
                changed = True
                if verbose:
                    print(f"❌ Removida variável {worst_feature} com p-valor {worst_pval:.3f}")

        if not changed:
            break

    if verbose:
        print("\n📊 Variáveis selecionadas pelo Stepwise:")
        print(included)
    return included


# ---------- 3.4 Aplicar a seleção stepwise ------------------

# Codificação e tratamento de tipos (garantir tudo como float)
df_encoded = pd.get_dummies(df, columns=cat_vars, drop_first=True)

# Converter todo o DataFrame para float
df_encoded = df_encoded.apply(pd.to_numeric, errors="coerce").astype(float)
df_encoded = df_encoded.fillna(df_encoded.mean())

# Executar stepwise selection
selected_vars = stepwise_selection(df_encoded, "G3", verbose=True)

# Criar modelo final com as variáveis selecionadas
X_final = sm.add_constant(df_encoded[selected_vars])
y_final = df_encoded["G3"]

model_final = sm.OLS(y_final, X_final).fit()

print("\n📈 Resumo do modelo final (Stepwise):")
print(model_final.summary())
```

# **Discussão Interpretativa**

### **Interpretação dos resultados**

As variáveis selecionadas pelo método *Stepwise* (**G2**, **absences**, **G1** e **age**) representam aquelas com maior poder explicativo sobre a variável dependente **G3 (nota final)**. O modelo final apresentou um **R² de 0,829** e um **R² ajustado de 0,827**, indicando excelente capacidade explicativa: aproximadamente **83% da variação** do desempenho final dos alunos pode ser atribuída a essas variáveis.

Os coeficientes estimados mostraram-se estatisticamente significativos (p < 0,05) e coerentes com a literatura. As notas anteriores (**G1** e **G2**) apresentaram coeficientes positivos, confirmando sua forte influência direta sobre o resultado final, em conformidade com **Cortez & Silva (2008)**, segundo os quais o desempenho prévio é o melhor preditor do sucesso acadêmico. Dentre elas, **G2** destacou-se como o fator mais determinante, sugerindo que o desempenho mais recente tem peso maior sobre o resultado final do ciclo.

A variável **age** apresentou coeficiente negativo e significativo (β = −0,19), indicando que alunos mais velhos tendem a obter desempenho ligeiramente inferior, possivelmente em função de fatores externos, como menor engajamento escolar ou trajetórias acadêmicas irregulares. Já a variável **absences** exibiu um coeficiente positivo inesperado (β = +0,04), o que pode ser resultado de **não linearidade**, **efeito de outliers** ou **correlação indireta** com outras variáveis (por exemplo, alunos que faltam mais podem também ter bom desempenho prévio, compensando ausências). Essa relação merece investigação complementar por meio de análises de resíduos ou modelos não lineares.

Apesar do bom ajuste global, testes de diagnóstico revelaram **violação dos pressupostos clássicos da regressão linear**. O **teste de Breusch–Pagan** indicou **heterocedasticidade significativa (p < 0,001)**, e o **teste de Jarque–Bera** evidenciou **não normalidade dos resíduos** (assimetria e alta curtose). Esses resultados sugerem que, embora os coeficientes estimados permaneçam não viesados, os **erros-padrão tradicionais podem estar sub ou superestimados**, afetando a validade dos testes *t* e dos intervalos de confiança. Para mitigar esse problema, recomenda-se a **utilização de erros-padrão robustos (HC3)** ou, alternativamente, **Weighted Least Squares (WLS)**, garantindo inferências mais confiáveis.

### **Limitações e considerações**

- Pode haver **multicolinearidade residual** entre variáveis fortemente correlacionadas, especialmente as notas intermediárias (**G1** e **G2**).
- Alguns preditores, como **consumo de álcool (Walc, Dalc)**, podem não ter sido selecionados pelo método *Stepwise* devido à baixa significância estatística no conjunto amostral, embora mantenham relevância teórica.
- **Dados autorreportados** (como hábitos de consumo e faltas) estão sujeitos a **viés de resposta**.
- Fatores **socioeconômicos, familiares e psicológicos** não mensurados podem exercer influência adicional sobre o desempenho.

- Conjunto **LINE** de verificação de modelo: Representando os quatro pilares de validade de um modelo de regressão linear múltipla:

    *   **L**inearidade: O pressuposto de Linearidade é atendido, pois as tendências evidenciadas nos scatterplots demonstram a ascedência entre os valores de G3xG1 e G3xG2.
    *   **I**ndependência de erros: O pressuposto de independência de erros é atendido, pois a base de dados individuais de estudantes não apresenta dependência sequencial. Uma evidência para tal pressuposto, é o valor de Durbin-Watson estar ~2 (1.866).
    *   **N**ormalidade: o pressupsoto de nornmalidade não é atendido totalmente, mesmo que a cauda à esquerda no gráfico de distribuição de resíduos seja de comum ocorrência, não comprometendo o modelo, o teste de Jarque–Bera evidenciou não normalidade dos resíduos (assimetria e alta curtose).
    *   Homocedasticidade ou **E**qual Variance: o pressuposto de homocedasticidade não é atendido, pois o teste de Breusch–Pagan indicou heterocedasticidade significativa (p < 0,001), sendo que o parâmetro para ter homocedasticidade seria p > 0.05.

---

### **Conclusão interpretativa**

Conclui-se que o **modelo linear proposto explica de forma robusta a variabilidade das notas finais (G3)**, tendo as notas anteriores e aspectos comportamentais como os principais preditores do desempenho escolar. Contudo, a detecção de heterocedasticidade e não normalidade dos resíduos indica que **modelos mais robustos** — como regressão robusta, transformações de variáveis ou modelos não lineares — podem aprimorar a precisão e a confiabilidade das inferências. Ainda assim, os resultados obtidos são consistentes com a literatura e confirmam a importância do desempenho histórico e da regularidade escolar como determinantes do sucesso acadêmico.
